```python
#Knowledge Base and the Userinterface

rules = {
    'is_bird': ['has_feathers','lays_eggs'],
    'is_fish': ['has_gills','has_fins','can_swim'],
    'is_mammal': ['has_fur','gives_live_birth'],
    'is_penguin': ['is_bird','cannot_fly','can_swim'],
    'is_bat': ['is_mammal','can_fly']
}

facts = {}

def get_user_fact(question):
    while True:
        answer = input(f"Does the animal {question}? (yes/no): ").lower().strip()
        if answer in ['yes', 'no']:
            facts[question] = (answer == 'yes')
            return facts[question]
        else:
            print("Please answer with 'yes' or 'no'.")
```

1) Examine the rules dictionary. WHat are the conditions for an animal to be classified as a 'is_bat'?

An animal is classified as bat if the condions 'is_mammal' and 'can_fly' become true.


2) If a user answers "yes" to "has_feathers", where is tis information stored? Whats is the purpose of the facts dictionary?

If the user answers "yes" to 'has_feathers', this information would be stored in the fact dictionary. The facts dictionary is for storing all the collected facts about the animals.




```python
#The Inference Engine – Backward Chaining
def backward_chaining(goal):
    if goal in facts:
        return facts[goal]

    if goal in rules:
        conditions = rules[goal]
        all_conditions_met = True
        for condition in conditions:
            if not backward_chaining(condition):
                all_conditions_met = False
                break
        if all_conditions_met:
            facts[goal] = True
            return True
        else:
            facts[goal] = False
            return False
    else:

        return get_user_fact(goal)

print("---Expert System: Animal Identifier----")
print("Please answer the following questions with 'yes' or 'no'.")

goal_animal = 'is_fish'
if backward_chaining(goal_animal):
    print(f"\nConclusion: The animal is a {goal_animal.split('_')[1]}!")
else:
    print(f"\nConclusion: The animal is not a {goal_animal.split('_')[1]}.")
```

1) Run the code and observe the order of the questions. Why does the system ask about "can_fly" before it asks about "has_feathers"?

It need to confirm whether it is a mammal or not. Backward chaining works depth-first: it tries to prove a goal by recursively checking its conditions in the order the rules list them.

Just need to change the goal animal into 'is_fish' in the code. The questions would be first 'has_gills',then 'has_fins' and finally 'can_swim'.

```python
def forward_chaining():
    print("\n---Forward Chaining Process---")


    initial_questions = ['can_fly', 'has_gills', 'lays_eggs']
    for question in initial_questions:
        get_user_fact(question)

    new_facts_added = True
    while new_facts_added:
        new_facts_added = False
        for goal, conditions in rules.items():
            if goal not in facts:
                all_conditions_met = True
                for condition in conditions:
                    if condition not in facts or not facts[condition]:
                        all_conditions_met = False
                        break
                if all_conditions_met:
                    facts[goal] = True
                    new_facts_added = True
                    print(f"Deduction: {goal} is true.")

    print("\n---Final Conclusion----")
    for fact, value in facts.items():
        if 'is_' in fact and value:
            print(f"The animal is a {fact.split('_')[1]}.")

# Run the forward chaining system
facts.clear()
forward_chaining()
```

1) Run the forward chaining code. Compare the experience to backward chaining. What are the key differences you notice?

Backward chaining starts with a specific goal, such as determining whether the animal is a penguin, and works backward through the rules to see if the necessary conditions are met. It asks questions only when needed to confirm the goal, making it more focused and efficient. Forward chaining, on the other hand, starts with known facts by asking initial questions and then iteratively applies rules to deduce all possible conclusions. This means it may ask questions that are not directly relevant to a specific goal, and the order of questions is more fixed, which can feel less efficient from the user’s perspective.

2) Based on your experience, which approach would be better for a medical diagnosis expert system? Why? (Hint: Think about which is more efficient for the user).

For a medical diagnosis system, backward chaining works better. Doctors usually start with a specific disease in mind, so the system can ask only the questions needed to confirm or rule out that disease. This saves time and avoids unnecessary questions. Forward chaining, on the other hand, asks many questions to figure out all possible conditions, which can overwhelm the patient with irrelevant queries.

```python
def run_expert_system_session():
  print("Agent (E5): Okay, activating the animal expert system.")
  print("agent (E5): Please answer a few questions (Yes/No).")
  facts.clear()
  possible_animals= ['is_bird','is_fish','is_mammal','is_bat']
  for animal_type in possible_animals:
    if backward_chaining(animal_type):
      print(f"\nAgent [E5]: The expert system concludes the animal is a {animal_type.split('_')[1]}!")
      print("\nAgent (ES): Expert System complete. Returning to chat.")
      return
  print("\nAgent (ES): The Expert system could not identify the animal.")
  print("Agent (ES): Expert system complete. Returing to chat.")
run_expert_system_session()
```

```python
# Part2: Add the UppBot chabtbot from Lab 3 Cont.

from textblob import TextBlob

class UppBot:
    def __init__(self, name="ElsaBot"):
        self.name = name
        self.agent_model = {
            'user_name': None,
            'greeted_before': False
        }

    def analyse_sentiment(self, text):
        analysis = TextBlob(text)
        if analysis.sentiment.polarity > 0.1:
            return "positive"
        elif analysis.sentiment.polarity < -0.1:
            return "negative"
        else:
            return "neutral"

    def get_response(self, user_input):
        user_input_lower = user_input.lower()
        user_sentiment = self.analyse_sentiment(user_input)

        print(f"Agent's sensor detects sentiment: {user_sentiment}")
        response = None  # initialize

        # Sentiment-based responses
        if user_sentiment == "negative":
            response = "I'm sorry to hear that. I hope things get better."
        elif user_sentiment == "positive":
            response = "That's great! I'm glad to hear that."

        # Fallback rule-based responses if no sentiment response
        if response is None:
            if "hello" in user_input_lower or "hi" in user_input_lower:
                if not self.agent_model['greeted_before']:
                    self.agent_model['greeted_before'] = True
                    response = "Hello there! How can I help you today?"
                else:
                    response = "Hi again! How can I assist you?"

            elif "how are you" in user_input_lower:
                if self.agent_model['user_name']:
                    response = f"I am doing great, thanks for asking, {self.agent_model['user_name']}!"
                else:
                    response = "I am a computer program, so I'm doing great!"

            elif "what is your name" in user_input_lower:
                response = f"My name is {self.name} and I am a chat agent."

            elif "my name is" in user_input_lower:
                name_start = user_input_lower.find("my name is") + len("my name is")
                user_name = user_input[name_start:].strip().title()
                self.agent_model['user_name'] = user_name
                response = f"Nice to meet you, {user_name}!"

            elif "goodbye" in user_input_lower or "bye" in user_input_lower:
                response = "Goodbye! Have a nice day."

            else:
                response = "I'm not sure how to respond to that."

        return response


chatbot = UppBot()
botname = chatbot.name

while True:
    try:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            print(f"{botname}: Hej då!")
            break
        response = chatbot.get_response(user_input)
        print(f"{botname}: {response}")
    except (KeyboardInterrupt, SystemExit):
        print(f"\n{botname}: Exiting... Hej då!")
        break
```

```python
#Set up the Expert System Chatbot

def expert_system_chat_agent():

    bot = UppBot(name="ElsaBot")

    print("Agent: Hello! I can chat or identify an animal.")
    print("Agent: Type 'identify an animal' to start the expert system, or 'exit' to quit.")

    while True:
        user = input("You: ")
        lower = user.lower().strip()

        if lower == "exit":
            print("Agent: Goodbye!")
            break

        if "identify an animal" in lower:
            run_expert_system_session()
            continue


        print(f"{bot.name}: {bot.get_response(user)}")



expert_system_chat_agent()
```

UppBot manages casual chat, greetings, and sentiment detection, while the expert system focuses on applying rules to identify animals. This separation of concerns makes the code easier to understand, maintain, and debug.
